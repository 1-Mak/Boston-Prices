```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
data = pd.read_csv('boston.csv')
display(data)
```

```python
plt.figure(figsize=(10, 5))
chart = sns.heatmap(data.corr(),annot=True, cmap='crest')
for i in chart.texts:
    if float(i.get_text()) >= 0.6 or float(i.get_text()) <= -0.6:
        i.set_text(i.get_text())
    else:
        i.set_text("")
```

```python
plt.figure(figsize=(10, 10))
sns.regplot(x='CRIM', y='MEDV', data=data, ci = None,  line_kws={'color': 'green'})
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.title('Цена дома от уровня преступности в районе')
plt.xlabel('Уровень преступности ')
plt.ylabel('Цена дома')
```

```python
plt.figure(figsize=(15, 10))
sns.regplot(x='RM', y='MEDV', data=data, ci = None,  line_kws={'color': 'green'})
plt.title('зависимость уровня агрессии человека от времени рядом с Егором')
plt.xlabel('Время в секундах ')
plt.ylabel('Уровень агрессии')
```

```python
plt.figure(figsize=(15, 10))
sns.regplot(x='NOX', y='MEDV', data=data, ci = None,  line_kws={'color': 'green'})

plt.title('Цена дома от в зависимости от загрязнения')
plt.xlabel('Загрязнение')
plt.ylabel('Цена дома')
```

```python
X = data[['CRIM']]
y = data['MEDV']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print("Результаты регрессии:")
print(model.summary())



```

```python
X = data[['RM']]
y = data['MEDV']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print("Результаты регрессии:")
print(model.summary())
```

```python
X = data.drop("MEDV", axis = 1)
y = data['MEDV']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print("Результаты регрессии:")
print(model.summary())
```

```python
X = data[['NOX']]
y = data['MEDV']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print("Результаты регрессии:")
print(model.summary())
```
